```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -12*x**4*np.sin(np.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
```

```python
roots = np.roots([-12, 0, -18, 5, 10, -30])
print("Корни функции f(x):", roots)
```

```python
x = np.linspace(-10, 10, 1000)  # Генерируем значения x на интервале [-10, 10]
y = f(x)  # Вычисляем значения функции f(x)

increasing_intervals = []
decreasing_intervals = []

for i in range(len(x) - 1):
    if y[i + 1] > y[i]:
        increasing_intervals.append((x[i], x[i + 1]))
    elif y[i + 1] < y[i]:
        decreasing_intervals.append((x[i], x[i + 1]))

print("Интервалы, на которых функция возрастает:", increasing_intervals)
print("Интервалы, на которых функция убывает:", decreasing_intervals)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x)')
plt.grid(True)
plt.show()
```

```python
vertex_x = x[np.argmin(y)]
vertex_y = np.min(y)
vertex = (vertex_x, vertex_y)
print("Вершина функции f(x):", vertex)
```

```python
positive_intervals = []

for i in range(len(x) - 1):
    if y[i] > 0:
        positive_intervals.append((x[i], x[i + 1]))

print("Промежутки, на которых f(x) > 0:", positive_intervals)
```

```python
negative_intervals = []

for i in range(len(x) - 1):
    if y[i] < 0:
        negative_intervals.append((x[i], x[i + 1]))

print("Промежутки, на которых f(x) < 0:", negative_intervals)
```
